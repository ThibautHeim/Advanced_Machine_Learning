```python
from sklearn.decomposition import PCA

dummy_data = torch.rand((100, 10))
pca = PCA(n_components=2)
pca.fit(dummy_data)
# Data on the new PCA space
transformed_data = pca.transform(dummy_data)
print(transformed_data.shape)
```
